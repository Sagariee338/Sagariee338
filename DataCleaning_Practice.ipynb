```python
nan_counts=df.isnull().sum()
print(nan_counts)
```

```python
nan_counts_sorted = nan_counts.sort_values(ascending=True)
print(nan_counts_sorted )
```

```python
nan_counts_sorted = nan_counts.sort_values(ascending=False)
print(nan_counts_sorted )
```

```python
df_filled = df.fillna(0)
print(df_filled)
```

```python
column_means = df.mean()
df.filled = df.fillna(column_means)
print(df_filled)
```

```python
column_std = df.std()
df.filled = df.fillna(column_std)
```

```python
print(df_filled)
```

```python
column_mode = df.mode()
df.filled = df.fillna(column_mode)
print(df_filled)
```
